### Cleaning Data

```python
# geolocation_df
# Hapus semua nilai duplicatenya
geolocation_df.drop_duplicates(inplace=True)

# cek nilai duplicate
print("Total duplicates = ", geolocation_df.duplicated().sum())
```

```python
# orders_df
# ubah tipe data ke date
date_cols = [
    'order_purchase_timestamp',
    'order_approved_at',
    'order_delivered_carrier_date',
    'order_delivered_customer_date' ,
    'order_estimated_delivery_date'
]

for col in date_cols:
  orders_df[col] = pd.to_datetime(orders_df[col])

# cek tipe data
orders_df.info()
```

```python
# Gabungkan orders_main_df dengan products_df
orders_product_df = pd.merge(
    left = orders_main_df,
    right= products_df,
    on='product_id',
    how='left'
)

orders_product_df.info()
# Hapus prder yang tidak masuk ke daftar product
orders_product_df.dropna(subset=['product_category_name'], inplace=True)

# cek product category yang tidak memiliki terjemahan in english
orders_product_df[orders_product_df['product_category_name_english'].isnull()]['product_category_name'].value_counts()

# mapping manual
mapping = {
    'pc_gamer' : 'Gaming_PC',
    'portateis_cozinha_e_preparadores_de_alimentos': 'kitchen_food_processors'
}

mask = orders_product_df['product_category_name_english'].isin(mapping.keys())
orders_product_df.loc[mask, 'product_category_name_english'] = orders_product_df.loc[mask, 'product_category_name'].map(mapping)

orders_product_df['product_category_name_english'].fillna('unknown', inplace=True)
```

## Analisis Lanjutan (Opsional)

### RMF Analysis

```python
"""RFM analysis dilakukan untuk mengelompokkan pelanggan
berdasarkan tiga kriteria utama. Analysis ini dapat digunakan untuk
menentukan strategi yang tepat bagi customer yang tepat"""

# Ambil tanggal terakhir sejak transaksi terkahir masuk
max_date = all_data['order_purchase_timestamp'].max() + pd.Timedelta(days=1)

# Hitung masing masing metrik RFM
rfm_df = all_data.groupby('customer_unique_id').agg({
    'order_purchase_timestamp':lambda x: (max_date - x.max()).days,
    'order_id':'nunique',
    'price':'sum'
})

# Ganti nama kolom
rfm_df.rename(columns = {
    'order_purchase_timestamp':'recency',
    'order_id':'frequency',
    'price':'monetary'
}, inplace=True)

rfm_df.sort_values(by='frequency', ascending=False).head(10)
```

```python
# Bagi customer untuk masing masing segment

# Recency score
rfm_df['r_score'] = pd.qcut(rfm_df['recency'], 5, labels=[5, 4, 3, 2, 1])

# Frequency score
rfm_df['f_score'] = pd.qcut(rfm_df['frequency'].rank(method='first'), 5, labels=[1, 2, 3, 4, 5])

# Gabung score menjadi satu string
rfm_df['rf_score'] = rfm_df['r_score'].astype(str) + rfm_df['f_score'].astype(str)

# Function untuk menempatkan segmentasi customer
def customer_segmentation(row):
  rf = row['rf_score']

  # aturan pemetaan
  if rf in ['55', '54', '45', '44']:
    return 'Champions'
  if rf in ['35', '34']:
    return 'Loyal customers'
  if rf in ['53', '52', '51', '43', '42', '41', '33', '32', '31']:
    return 'Potential loyalist'
  if rf in ['25', '24', '15', '14']:
    return 'At risk'
  if rf in ['33', '23', '22', '32']:
    return 'Hibernating'
  else:
    return 'Lost'

rfm_df['customers_type'] = rfm_df.apply(customer_segmentation, axis = 1)
rfm_df.sort_values(by='recency', ascending = True).head(10)
```

```python
# visualisasi untuk mengetahui sebaran customer pada tiap segment

customer_type = ['Champions', 'Loyal customers', 'Potential loyalist', 'At risk', 'Hibernating', 'Lost']
rfm_clean = rfm_df[rfm_df['customers_type'].isin(customer_type)]

fig, axes = plt.subplots(1, 2, figsize = (18, 6))
count_data = rfm_clean['customers_type'].value_counts().reindex(customer_type)

axes[0].pie (
  count_data,
  labels = count_data.index,
  autopct = '%1.1f%%',
  startangle = 140,
  colors = sns.color_palette('viridis', len(count_data)),
)
axes[0].set_title('Customer Type Distribution', weight='bold')

# Visualisasi monetary
revenue_per_type = rfm_clean.groupby('customers_type')['monetary'].sum().reindex(customer_type)

sns.barplot(x=revenue_per_type.values, y=revenue_per_type.index, palette = 'magma', hue=revenue_per_type.index, legend = False, ax=axes[1])
axes[1].set_title('Customer Type by Revenue', weight='bold')
axes[1].set_xlabel('Total Revenue')
axes[1].xaxis.set_major_formatter(FuncFormatter(millions_formatter))
axes[1].bar_label(axes[1].containers[0], fmt=lambda x: f'{x/1e6:.1f}M', padding=3)

plt.tight_layout()
plt.show()
```
